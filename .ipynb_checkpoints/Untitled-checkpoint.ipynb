```python
import pandas as pd
import yfinance as yf
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
import matplotlib.pyplot as plt
```

```python
meta = pd.read_csv('symbols_valid_meta.csv')
meta = meta.dropna(subset=['Symbol'])
```

```python
stock_symbol = meta[meta['ETF'] == 'N']['Symbol'].iloc[0]
print("Using symbol:", stock_symbol)

df = yf.download(stock_symbol, start="2020-01-01", end="2023-12-31")

df['Next_Close'] = df['Close'].shift(-1)
df = df.dropna()

X = df[['Open', 'High', 'Low', 'Close', 'Volume']]
y = df['Next_Close']

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, shuffle=False)


model = LinearRegression()
model.fit(X_train, y_train)

y_pred = model.predict(X_test)

print("MSE:", mean_squared_error(y_test, y_pred))
plt.figure(figsize=(10,5))
plt.plot(y_test.values, label='Actual')
plt.plot(y_pred, label='Predicted')
plt.legend()
plt.title(f"{stock_symbol} - Close Price Prediction")
plt.show()
```

```python
from sklearn.metrics import r2_score, mean_squared_error

# Predict on training data
y_train_pred = model.predict(X_train)

# R2 Score
r2 = r2_score(y_train, y_train_pred)
print(f"Training R Score (Accuracy): {r2:.4f}")

# Optional: also print RMSE
rmse = mean_squared_error(y_train, y_train_pred, squared=False)
print(f"Training RMSE: {rmse:.4f}")
```

```python
# Predict on test data
y_test_pred = model.predict(X_test)
r2_test = r2_score(y_test, y_test_pred)
print(f"Test R Score (Accuracy): {r2_test:.4f}")
```

```python
# Did we train it?
model.fit(X_train, y_train)   #  Yes

# Predict training and test sets
train_preds = model.predict(X_train)
test_preds = model.predict(X_test)

# Evaluate
from sklearn.metrics import r2_score, mean_squared_error
print("Train R2:", r2_score(y_train, train_preds))
print("Test R2:", r2_score(y_test, test_preds))
```

```python
# Get the last row of your feature set
latest_features = X.iloc[[-1]]  # double brackets keep it 2D

# Predict
next_day_price = model.predict(latest_features)
print(f"Predicted Next Day Closing Price: {next_day_price[0]:.2f}")
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10,5))
plt.plot(y_test.values, label="Actual", linewidth=2)
plt.plot(test_preds, label="Predicted", linestyle='--')
plt.title("Actual vs Predicted Stock Prices (Test Set)")
plt.xlabel("Time Index")
plt.ylabel("Price")
plt.legend()
plt.grid(True)
plt.show()
```

```python
import joblib
joblib.dump(model, 'stock_price_model.pkl')
```
